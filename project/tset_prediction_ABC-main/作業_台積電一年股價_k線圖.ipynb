```python
import pandas as pd
import yfinance as yt
from datetime import datetime
```

```python
# 取得歷史股價
today = datetime.today().strftime('%Y-%m-%d')
df = yt.download('2330.TW', start='2022-09-03', end=today)
df.tail(10)
```

```python
# Importing the necessary charting library
import matplotlib.pyplot as plt
# The syntax to plot a line chart

df2 = df

plt.figure(figsize=(14, 8))
plt.plot(df2.index, df2.Close, color = 'black')
# The syntax to add the label created above
plt.legend()
# The syntax to add a grid
plt.grid()
```

```python
def ohlc_plot(Data, name):
    BAR_WIDTH = 8
    plt.figure(figsize=(14, 8))
    for i in range(len(Data)): 
        plt.vlines(x = Data.index[i], ymin = Data.iloc[i, 2], ymax = Data.iloc[i, 1], 
                   color = 'black', linewidth = 1)

        if Data.iloc[i, 0] > Data.iloc[i, 3]:
            color_Data = 'green'
            plt.vlines(x = Data.index[i], ymin = Data.iloc[i, 3], ymax = Data.iloc[i, 0], 
                       color = color_Data, linewidth = BAR_WIDTH)              
        elif Data.iloc[i, 0] < Data.iloc[i, 3]:
            color_Data = 'red'
            plt.vlines(x = Data.index[i], ymin = Data.iloc[i, 0], ymax = Data.iloc[i, 3], 
                       color = color_Data, linewidth = BAR_WIDTH)  
        else:
            color_Data = 'black'
            plt.vlines(x = Data.index[i], ymin = Data.iloc[i, 0], ymax = Data.iloc[i, 3], 
                       color = color_Data, linewidth = BAR_WIDTH)  
          
    plt.grid()
    plt.title(name)
    
# Using the function
ohlc_plot(df, '')
```
